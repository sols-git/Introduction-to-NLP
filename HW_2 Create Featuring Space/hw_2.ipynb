4.Проверьте ваши векторайзеры на корпусе который использовали на вебинаре, составьте таблицу метод векторизации и скор который вы получили (в методах векторизации по изменяйте параметры что бы добиться лучшего скора) обратите внимание как падает/растёт скор при уменьшении количества фичей, и изменении параметров, так же попробуйте применить к векторайзерам PCA для сокращения размерности посмотрите на качество сделайте выводы

```python
# Загружаем данные
data = open('corpus').read()
labels, texts = [], []
for i, line in enumerate(data.split("\n")):
    content = line.split()
    labels.append(content[0])
    texts.append(" ".join(content[1:]))

# создаем df
trainDF = pd.DataFrame()
trainDF['text'] = texts
trainDF['label'] = labels
trainDF.head(5)
```

```python
from sklearn import model_selection, preprocessing, linear_model
from sklearn.metrics import accuracy_score
train_x, valid_x, train_y, valid_y = model_selection.train_test_split(trainDF['text'], trainDF['label'])

# labelEncode целевую переменную
encoder = preprocessing.LabelEncoder()
train_y = encoder.fit_transform(train_y)
valid_y = encoder.fit_transform(valid_y)


count_vect = CountVectorizer(analyzer='word', token_pattern=r'\w{1,}')
count_vect.fit(trainDF['text'])

xtrain_count =  count_vect.transform(train_x)
xvalid_count =  count_vect.transform(valid_x)

classifier = linear_model.LogisticRegression()
classifier.fit(xtrain_count, train_y)
predictions = classifier.predict(xvalid_count)
predictions
```

```python
def investigation_scor_vectorasers(max_features = 1000, max_df=0.9):
    count_vectorizer = CountVectorizer(ngram_range=(1, 1), analyzer='word', max_df = max_df, max_features = max_features, stop_words='english', lowercase=False)
    tfidf_vectorizer = TfidfVectorizer(max_df=max_df,max_features=max_features,stop_words='english')
    hashing_vectorizer = HashingVectorizer(n_features=max_features, stop_words='english')

    vectorizers = [count_vectorizer, tfidf_vectorizer, hashing_vectorizer]
    scor_vs_vectorasers = []

    for v in vectorizers:
        v.fit(trainDF['text'])
        xtrain_count =  v.transform(train_x)
        xvalid_count =  v.transform(valid_x)

        classifier = linear_model.LogisticRegression()
        classifier.fit(xtrain_count, train_y)
        predictions = classifier.predict(xvalid_count)
        scor_vs_vectorasers.append([v, accuracy_score(valid_y, predictions)])
    return pd.DataFrame(scor_vs_vectorasers, columns = ['Vectoraser', 'Score'])
```

```python
investigation_scor_vectorasers(max_features = 1000, max_df=0.9)
```

```python
investigation_scor_vectorasers(max_features = 5000, max_df=0.9)
```

```python
investigation_scor_vectorasers(max_features = 1000, max_df=0.6)
```

```python
investigation_scor_vectorasers(max_features = 1000, max_df=1)
```
